### Model
#### AutoEncoder
AutoEncoder의 목적은 $x$와 $f(x) = decode(encode(x))$ 간의 error를 최소화 하는 data의 **d-dimensional representation**을 찾는 것이다. 이를 구현하는 AutoEncoder의 구조는 크게 다음과 같다.

1. Encode: $R^n \rightarrow R^d$
2. Decode: $R^d \rightarrow R^n$


* 가운데의 z layer는 **code**라고 불린다.

#### decoder - mirror - encoder
Decoder mirrors encoder architecture. So decoder's weight $W^l_d$ is equal to transposed encoder weight $W^l_e$.

By decoder mirroring encoder, we can have free parameters two time less than uncostrained one.

#### Forward pass and inference
* input: partially observed user vector - very **sparse vector**
* output: **dense vector**

### Loss function
Actual rating이 없는 item을 평가에 사용 할 수는 없다. 따라서, **Masked Mean Squared Error** loss를 사용한다.

$$ MMSE = {m_i * (r_i - y_i)^2 \over \sum_{i=0}^{i=n} m_i} $$

where $r_i$ is acutal rating, $y_i$ is reconstructed rating,  and $m_i$ is a mask  function such that $m_i = 1$ if $r_i != 0$ else $m_i = 0$

### Dense re-feeding
ideal $f$가 존재한다고 가정해보면, 
* $f(x)_i$는 item i에 대한 모든 user의 future rating을 정확하게 예측함.
* 어떤 user가 새로운 item k를 평가한다면, $f(x')_k = x'_k$ 이고 $f(x) = f(x')$이 된다.

$\Rightarrow$ 즉, $y=f(x)$는 autoencoder의 **fixed point**여야 한다.

#### Training
**Fixed Point** 제약과 함께 dense  training  update를 실행하기 위해 다음과 같은 training step을 거친다.
1. sparse vector x가 주어지면, dense vector인 $f(x)$와 loss를 계산한다. (**Forward Pass)
2. gradient를 계산하고 weight update를 진행한다. (**Backward pass**)
3. $f(x)$를 새로운 input으로 취급하여, $f(f(x))$와 loss를 계산한다. 이 때, $f(x), f(f(x)), loss$의 값은 모두 non-zero이다.

### Experiment Setup
#### Train-Test Split
현실적으로 생각한다면, 우리가 추천시스템을 적용한다는 것은 *과거*의 데이터를 가지고, *미래*의 평점을 예측하는 것이다. 따라서, 주어진 data를 가지고  model을 설계할 때에, **시간**을 기준으로 data를 split해야 한다.
* Train set: 특정 timestamp 전의 data
* Test set(Test + Validation): 특정 timestamp 이후의 data

#### user-item 제외
training set에 등장하지 않는 user와 item은 test set과 validation set에서도 제외해야 한다.
